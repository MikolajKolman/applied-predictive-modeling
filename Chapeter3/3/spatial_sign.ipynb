```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal
from sklearn.preprocessing import normalize, scale
from IPython.core.debugger import set_trace
```

```python
data = multivariate_normal.rvs(mean=(3, 2), size=1000, cov=[
    [1, 0.6],
    [0.6, 1]
])
plt.scatter(data[:, 0], data[:, 1])
```

```python
data_spatial_signed = normalize(data, norm='l2')
plt.scatter(data_spatial_signed[:, 0], data_spatial_signed[:, 1])
```

```python
data_scaled = scale(data)
plt.scatter(data_scaled[:, 0], data_scaled[:, 1])
```

```python
data_spatial_signed = normalize(data_scaled, norm='l2')
plt.scatter(data_spatial_signed[:, 0], data_spatial_signed[:, 1])
```

```python
def spatial_sign(x):
#     set_trace()
    denom = np.sum(x ** 2)
    return x / np.sqrt(denom)
```

```python
data_spatial_signed = spatial_sign(data_scaled)
plt.scatter(data_spatial_signed[:, 0], data_spatial_signed[:, 1])
```
